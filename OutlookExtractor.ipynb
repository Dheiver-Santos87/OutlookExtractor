```python
import os
import win32com.client

class OutlookExtractor:
  def __init__(self):
    # Crie uma conexÃ£o com o Outlook
    self.outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")
    # Selecione a pasta de entrada (onde os emails sÃ£o armazenados)
    self.inbox = self.outlook.GetDefaultFolder(6)

  def extract_attachments(self):
    # Armazene a quantidade de emails na pasta de entrada
    initial_count = self.inbox.Items.Count
    # Crie a pasta de destino (em seus downloads) se ela nÃ£o existir
    root_folder = os.path.join(os.environ['USERPROFILE'], "Downloads", "Anexos")
    if not os.path.exists(root_folder):
      os.makedirs(root_folder)
      print("Pasta de destino criada")
    # Inicialize a contagem de emails processados
    processed_count = 0
    # Percorra cada email
    for message in self.inbox.Items:
      print(f"Processando email {processed_count + 1}: {message.Subject}")
      # Percorra cada anexo do email
      for attachment in message.Attachments:
        print(f"Processando anexo: {attachment.FileName}")
        # Obtenha a extensÃ£o do anexo
        extension = os.path.splitext(attachment.FileName)[1]
        # Crie a pasta de destino de acordo com a extensÃ£o do anexo
        subfolder = os.path.join(root_folder, extension)
        if not os.path.exists(subfolder):
          os.makedirs(subfolder)
          print(f"Pasta {extension} criada")
        # Salve o anexo na pasta de destino correspondente
        attachment.SaveAsFile(os.path.join(subfolder, attachment.FileName))
        print(f"Anexo salvo em {os.path.join(subfolder, attachment.FileName)}")
      processed_count += 1
    # Exiba a quantidade total de emails processados
    print(f"{processed_count} emails processados de um total de {initial_count}")

# Crie uma instÃ¢ncia da classe OutlookExtractor
extractor = OutlookExtractor()
# Extraia os anexos dos emails
extractor.extract_attachments()
```
